```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Datasets\Salary.csv')
```

```python
df.head()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y, color = 'red')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred.astype(int)})
df.head(20)
```

```python
df_1 = df.head(25)
df_1.plot(kind='bar',figsize=(8,5))
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')
plt.show()
```

```python
plt.scatter(X, y, color='blue', label='Data Points')
plt.plot(X, regressor.predict(X), color='red', label='Regression Line')
for i in range(len(X)):
    plt.vlines(x=X[i], ymin=y[i], ymax=regressor.predict(X)[i], color='green', linestyle='dotted')
plt.title('Simple Linear Regression')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, y_pred)

print(f"R-squared: {r2:.4f}")
```
